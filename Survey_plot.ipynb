```python
import pickle

with open(path+'QSANN_RP_wo_pos_enco.pkl', "rb") as input_file:
    RP_data = pickle.load(input_file)
```

```python
MC_data.keys()
```

```python
epoch=np.linspace(0,199,200)
RP_epoch=np.linspace(0,299,300)
```

```python
fig,axes = plt.subplots(2,2,sharex = False, sharey = False, figsize = (1.35*3.375,1.35*0.85*3.375),squeeze = False)
plt.subplots_adjust(wspace=0.05, hspace=0.15)

axes[0,0].plot(epoch,MC_data['Test_Acc'],c='r',label='$\mathrm{test}$')
axes[0,0].plot(epoch,MC_data['Train_Acc'],c='b',label='$\mathrm{train}$')

#axes[0,0].set_xlabel('$epoch$')
axes[0,0].set_ylabel('$\mathrm{Accuracy}$')
axes[0,0].yaxis.set_label_coords(-0.25,0.45)
#axes[0,0].legend(fontsize=20,frameon=False)
axes[0,0].legend(frameon=False)
axes[0,0].set_ylim(0.45,1.0)
axes[0,0].text(-0.17,1.05,r"$\mathrm{(a)}$",transform=axes[0,0].transAxes) # To adjust the position of text beside axis
axes[0,0].set_xticks([])

axes[0,1].plot(RP_epoch,RP_data['Test_Acc'],c='r')
axes[0,1].plot(RP_epoch,RP_data['Train_Acc'],c='b')

axes[0,1].set_yticks([]) # To make y label disappear
axes[0,1].set_xticks([])
axes[0,1].set_ylim(0.45,1.0)


axes[0,1].text(-0.05,1.05,r"$\mathrm{(b)}$",transform=axes[0,1].transAxes)

axes[1,0].plot(epoch,MC_data['Test_loss'],c='r')
axes[1,0].plot(epoch,MC_data['Train_loss'],c='b')

axes[1,0].set_xlabel('$\mathrm{Epoch}$')
axes[1,0].set_ylabel('$\mathrm{Loss}$')
axes[1,0].yaxis.set_label_coords(-0.2,0.45)
axes[1,0].set_ylim(0.0,1.55)
#axes[0,2].legend(fontsize=18,frameon=False)
axes[1,0].legend(frameon=False)
axes[1,0].text(-0.17,1.04,r"$\mathrm{(c)}$",transform=axes[1,0].transAxes)


axes[1,1].plot(RP_epoch,RP_data['Test_loss'],c='r')
axes[1,1].plot(RP_epoch,RP_data['Train_loss'],c='b')

#axes[1,1].set_yscale('log')
#axes[1,1].set_xscale('log')
axes[1,1].set_yticks([])
axes[1,1].set_xlabel('$\mathrm{Epoch}$')
#axes[1,1].yaxis.set_label_coords(-0.15,0.42)
axes[1,1].legend(frameon=False,loc=4)

axes[1,1].text(-0.05,1.04,r"$\mathrm{(d)}$",transform=axes[1,1].transAxes)
plt.savefig("MC_RP_WO_pos_eco.png",dpi = 1000,bbox_inches='tight',facecolor='white',transparent=False)
```
